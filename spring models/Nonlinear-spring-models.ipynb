```python
import numpy as np

import matplotlib.pyplot as plt
```

```python
def exponential_history(deform, history):
    
    if deform >= history:
        
        traction = exponential(deform)
        
    else:
        
        traction = deform * (exponential(history) / history)
    
    return traction
```

```python
def exponential(deform):

    traction = 4 * deform * np.exp(-4*deform)
    
    return traction
```

```python
deform = np.linspace(0,0.5,100) # the deformation is assumed ranging from 0 to 0.5
```

```python
exponential_history_record = []

history = 0.25 # the maximum stretch is assumed as 0.25

for i in deform:
    
    exponential_history_record.append(exponential_history(i,history))
```

# Note: different spring models:

(1) linear model $F_1(x) = -kx$

(2) quadratic mode $F_2(x) = -kx^2$

(3) exponential model $F_3(x) = 4x \exp(-4x)$

(4) exponential model with history variable $F_4(x) = x \frac{F_3(x_h)}{x_h}$

where $x$ is the current deformation and $x_h$ is the history variable that equals the maximum deformation during the previous computation. Introduction of this history variable enables damage accumulation in springs. Specifically, the area between the blue curve (exponential model) and the orange curve (exponential model with history variable) represents the energy dissipation.

```python
# Visualization for differences between linear model and nonlinear models, including the quadratic model, the exponential model, and the exponential model with history variable

plt.figure(figsize=(10,10))

plt.plot(deform, exponential(deform))

plt.plot(deform,exponential_history_record)

plt.plot(deform, deform)

plt.plot(deform, deform**2)

plt.legend(['exponential','history','linear','quadratic'])

plt.ylim(0,0.5)

plt.xlim(0,0.5)

plt.show()
```
